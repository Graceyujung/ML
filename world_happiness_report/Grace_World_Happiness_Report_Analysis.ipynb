```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data[column] < lower_bound) | (data[column] > upper_bound)]

numerical_cols = final_df.select_dtypes(include=['number']).columns
for col in numerical_cols:
    outliers = detect_outliers_iqr(final_df, col)
    print(f"Outliers in {col}: {len(outliers)}")
```

```python
plt.figure(figsize=(16,10))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(3, 5, i)
    sns.histplot(final_df[col], kde=True, bins=20, color='green')
    plt.title(f'Histogram of {col}')
    plt.tight_layout()
plt.show()
```

Observation: Since the dataset has fewer outliers and I want to keep all data, imputation might be a reasonable approach to handle missing values.

```python
final_df['Country'] = final_df['Country'].replace({'Hong Kong S.A.R., China' : 'Hong Kong'})
final_df['Country'] = final_df['Country'].replace({'North Cyprus' : 'Northern Cyprus'})
final_df['Country'] = final_df['Country'].replace({'Somaliland region' : 'Somaliland Region'})
final_df['Country'] = final_df['Country'].replace({'Taiwan Province of China' : 'Taiwan'})
final_df['Country'] = final_df['Country'].replace({'Trinidad & Tobago':'Trinidad and Tobago'})
sorted(final_df['Country'].unique())
```

Observation: Similar country names have been standardized to a single, unified name.

```python
final_df['Region'].unique()
```

```python
country_region = final_df.dropna(subset=['Region'])[['Country', 'Region']].set_index('Country')['Region'].to_dict()

final_df['Region'] = final_df.apply(
    lambda row: country_region.get(row['Country'], row['Region']), axis=1)

final_df[final_df['Region'].isna()]
```

```python
final_df.loc[final_df['Country'] == 'North Macedonia', 'Region'] = 'Southeast Europe'

# the Gambia is divided into six regions
final_df.loc[final_df['Country'] == 'Gambia', 'Region'] = 'Unknown (Gambia)'

final_df.isnull().sum()
```

Observation: After applying a lambda function for Region NaNs, only two values were missing, so imputed manually.

```python
# Standard Error = (Upper CI - Lower CI) / (2 * z)
z = 1.96
mask = (final_df['Standard Error'].isna() &
        final_df['Lower Confidence Interval'].notna() &
        final_df['Upper Confidence Interval'].notna())

final_df.loc[mask, 'Standard Error'] = (
    final_df.loc[mask, 'Upper Confidence Interval'] - final_df.loc[mask, 'Lower Confidence Interval']) / (2 * z)
    

final_df.isnull().sum()
```

Observation: 312 of 624 missing values are successfully filled in.

```python
# Lower CI = Mean - z * Standard Error
# Upper CI = Mean + z * Standard Error

mask = (
    final_df['Standard Error'].notna() &
    final_df['Happiness Score'].notna() &
    final_df['Lower Confidence Interval'].isna() &
    final_df['Upper Confidence Interval'].isna())

final_df.loc[mask, 'Lower Confidence Interval'] = final_df.loc[mask, 'Happiness Score'] - z * final_df.loc[mask, 'Standard Error']
final_df.loc[mask, 'Upper Confidence Interval'] = final_df.loc[mask, 'Happiness Score'] + z * final_df.loc[mask, 'Standard Error']

final_df.isnull().sum()
```

Observation: Available 'Standard Error' and 'Happiness Score' filled in all the missing confidence intervals.

```python
final_df['Standard Error'] = final_df['Standard Error'].fillna(final_df['Standard Error'].median())
final_df['Lower Confidence Interval'] = final_df['Lower Confidence Interval'].fillna(final_df['Lower Confidence Interval'].mean())
final_df['Upper Confidence Interval'] = final_df['Upper Confidence Interval'].fillna(final_df['Upper Confidence Interval'].mean())

final_df.isnull().sum()
```

Observation: Since 'Standard Error' histogram is a bit skewed with 10 outliers, median imputation is applied, on the other hand, both CI has symmetric histogram with no outlier, so meam imputation is used for missing values.

```python
final_df['Trust (Government Corruption)'] = final_df['Trust (Government Corruption)'].fillna(final_df['Trust (Government Corruption)'].median())
final_df.isnull().sum()
```

Observation: 'Trust' has the most outliers with skewed histogram, median imputation has been applied to one missing value

```python
final_df['Dystopia Residual'] = final_df['Dystopia Residual'].fillna(final_df['Dystopia Residual'].mean())
final_df.isnull().sum()
```

Observation: The distribution is symmetric and outliers are few and not extreme, using the mean is fine.
